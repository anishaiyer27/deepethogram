This notebook is meant to provide a framework to preprocess lab data taken as mouse behavior videos from the lab camera into the format required for DeepEthogram, an automatic behavior scoring software. DeepEthogram is trained on infrared data, so this Notebook will aim to augment mouse videos to look like they have been taken with an infrared camera.

```python
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
import numpy as np
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pylab
import imageio
```

Regardless of how many times you run this cell, you will only get 3 additional pictures (plus the original one you loaded) because the filenames overwrite themselves.

```python
"""
Open and read the image into a python variable. This image was screenshotted
from one of Ricky's behavior videos and uploaded to Jupyter Notebook.
"""
im = Image.open("mouse1.png")

"""
Creates an object which we can apply functions on to ennhance the brightness
and contrast of the image with built-in functions.
"""
enhancer = ImageEnhance.Contrast(im)

"""
Enhance the contrast using the above enhancer object (variable) by some factor.
Enhancing by a factor of 1 will give us the original image. By calling the save
function on the output with title "original-image.png", we save a new image
under the title "original-image.png", which should be accessible in the Jupyter
Notebook file dropdown that got us to this tab AND locally wherever the GitHub
folder is on your computer.
"""
factor = 1 #gives original image
im_output = enhancer.enhance(factor)
im_output.save('original-image.png')

"""
Enhancing the contrast by a factor less than one will decrease the contrast.
"""
factor = 0.5 #decrease constrast
im_output = enhancer.enhance(factor)
im_output.save('less-contrast-image.png')

"""
Enhancing the contrast by a factor greater than one will increase the contrast.
"""
factor = 1.5 #increase contrast
im_output = enhancer.enhance(factor)
im_output.save('more-contrast-image.png')
```

```python
original = Image.open('mouse1.png')
original_enh = Image.open('original-image.png')
less_contrast = Image.open('less-contrast-image.png')
more_contrast = Image.open('more-contrast-image.png')

display("original", original)
display("original, enhanced by factor of 1 (should be same as original)", original_enh)
display("less contrast, enhanced by factor of 0.5", less_contrast)
display("more contrast, enhanced by factor of 1.5", more_contrast)
```

```python
def gray(filename):
    image = cv2.imread(filename)
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    plt.imshow(gray, cmap='gray')
    plt.show()
gray('mouse1.png')
gray('original-image.png')
gray('less-contrast-image.png')
gray('more-contrast-image.png')
```

```python
"""
This function applies this formula to convert this image to grayscale based
on its red, green, and blue components. It also displays the image.
Formula: gray = 0.2989 * R + 0.5870 * G + 0.1140 * B
"""
def rgb_to_grayscale(rgb):
    gray = np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
    plt.imshow(gray, cmap=plt.get_cmap('gray'), vmin=0, vmax=1)
    plt.show()

img = mpimg.imread('mouse1.png')
img_original = mpimg.imread('original-image.png')
img_less_contrast = mpimg.imread('less-contrast-image.png')
img_more_contrast = mpimg.imread('more-contrast-image.png')

rgb_to_grayscale(img)
rgb_to_grayscale(img_original)
rgb_to_grayscale(img_less_contrast)
rgb_to_grayscale(img_more_contrast)
```

## Loading in Data to Process

```python
# trying to convert frames of an mp4 file to grayscale
filename = 'pos_hist.mp4'
vid = imageio.get_reader(filename,  'ffmpeg')
nums = np.arange(20)
images = []
for num in nums:
    image = vid.get_data(num)
    fig = pylab.figure()
    fig.suptitle('image #{}'.format(num), fontsize=20)
    rgb_to_grayscale(image)
    #pylab.imshow(image)
pylab.show()
```

```python
vid_capture = cv2.VideoCapture('pos_hist.mp4')
if (vid_capture.isOpened() == False):
    print("Error opening the video file")
else:
    # Get frame rate information
    fps = int(vid_capture.get(5))
    print("Frame Rate : ",fps,"frames per second") 
    # Get frame count
    frame_count = vid_capture.get(7)
    print("Frame count : ", frame_count)
```

```python
success,image = vid_capture.read()
count = 0
while success:
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY) 
    cv2.imwrite("frame%d.png" % count, gray)     # save frame as png file      
    success,image = vid_capture.read()
    print('Read a new frame: ', success)
    count += 1
```

```python
fourcc = cv2.VideoWriter_fourcc(*'mp4v') 
video = cv2.VideoWriter('video.mp4', fourcc, 1, (20, 20))

for j in range(count):
    img = cv2.imread("frame%d.png" % + j)
    video.write(img)

cv2.destroyAllWindows()
video.release()
```
